```python
import cv2
import os as OS
import numpy as np
import matplotlib.pyplot as plt

class Img_Manipulation():
    def __init__(self, image_path):
        self._imagePath = image_path
        self._imageContainer = cv2.imread(self._imagePath)
        
    def img_border(self, width):
        print(width)
        Imgheight, Imglength = self._imageContainer.shape[:2]
        for height in range(Imgheight):
            for length in range(Imglength):
                if height < width or height >= Imgheight - width:
                    self._imageContainer[height][length] = 255
                
                elif height > width and length < width or height > width and length >= Imglength - width:
                    self._imageContainer[height][length] = 255
        
        plt.imshow(self._imageContainer)
        
    def img_rotate(self, degree):
        Imgheight, Imglength = self._imageContainer.shape[:2]
        rotated_matrix = cv2.getRotationMatrix2D((Imgheight / 2, Imglength / 2), degree, 1)
        rotated_image = cv2.warpAffine(self._imageContainer, rotated_matrix, (Imgheight, Imglength))
        plt.imshow(rotated_image)
    
    def img_scale(self, scale):
        scaleX, scaleY = scale.split(",")[0], scale.split(",")[1]
        scaled_image = cv2.resize(self._imageContainer, None, fx=int(scaleX), fy=int(scaleY), interpolation=cv2.INTER_CUBIC)
        plt.imshow(scaled_image)
    
    def img_perspective(self, skew_choice):
        Imgheight, Imgwidth = self._imageContainer.shape[:2]
        print(Imgheight, Imgwidth)
        print(skew_choice)
        source_points = np.float32([[0,0], [Imgwidth - 1, 0], [0,Imgheight - 1], [Imgwidth - 1, Imgheight - 1]])
        print(source_points)
        if skew_choice == "1":
            destination_points = np.float32([[0.1* Imgwidth,0], [0.9*Imgwidth, 0],[0,Imgheight - 1], [Imgwidth - 1, Imgheight - 1]])
            project_matrix = cv2.getPerspectiveTransform(source_points, destination_points)
            img_out = cv2.warpPerspective(self._imageContainer, project_matrix, (Imgheight, Imgwidth))
            plt.imshow(img_out)
    
    def img_translation(self, translate):
        Imgheight, Imgwidth = self._imageContainer.shape[:2]
        translateX , translateY = translate.split(",")[0], translate.split(",")[1]
        print(translateX, translateY)
        translate_matrix = np.float32([[1, 0, int(translateX)], [0, 1, int(translateY)]])
        translated_image = cv2.warpAffine(self._imageContainer, translate_matrix, (Imgwidth, Imgheight))
        plt.imshow(translated_image)
        
    def grey_img(self):
        grey_image = cv2.imread(self._imagePath, 0)
        plt.imshow(grey_image)
    
    def blue_img(self):
        Imgheight, Imgwidth = self._imageContainer.shape[:2]
        Blue,Green,Red=cv2.split(self._imageContainer)
        zeros=np.zeros((Imgheight,Imgwidth),dtype="uint8")
        cv2.imshow("Blue",cv2.merge([Blue,zeros,zeros]))
        blue_image = cv2.merge([Blue,zeros,zeros])
        plt.imshow(blue_image)
    
    def green_img(self):
        Imgheight, Imgwidth = self._imageContainer.shape[:2]
        Blue,Green,Red=cv2.split(self._imageContainer)
        zeros=np.zeros((Imgheight,Imgwidth),dtype="uint8")
        green_image = cv2.merge([zeros,Green,zeros])
        plt.imshow(green_image)
        
    def red_img(self):
        Imgheight, Imgwidth = self._imageContainer.shape[:2]
        Blue,Green,Red=cv2.split(self._imageContainer)
        zeros=np.zeros((Imgheight,Imgwidth),dtype="uint8")
        red_image = cv2.merge([zeros,zeros,Red])
        plt.imshow(red_image)
        
    def merge_image(self, second_image):
        second_image = cv2.imread(second_image)
        merged_image = cv2.add(self._imageContainer, second_image)
        cv2.imshow("Merge_Image", merged_image)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
    
    def blend_image(self, second_image):
        second_image = cv2.imread(second_image)
        blended_image = cv2.addWeighted(self._imageContainer, 0.7, second_image, 0.3, 0)
        cv2.imshow("blended_image", blended_image)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
        
    def img_threshold(self):
        grey_image = cv2.imread(self._imagePath, 0)
        retval, threshold = cv2.threshold(grey_image,127,255,cv2.THRESH_BINARY)
        cv2.imshow("threshold", threshold)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
        
    def img_blur(self, method_number):
        if method_number == "1":
            blurred_image = cv2.blur(self._imageContainer,(5,5))
        elif method_number == "2":
            blurred_image = cv2.GaussianBlur(self._imageContainer,(5,5),0)
        plt.imshow(blurred_image)
        
            
def main():
    path = input("Enter Path For Your Image: ")
    # C:\Users\Anurag\Desktop\OpenCvProject\404.jpg
    imgObj = Img_Manipulation(path)
    print("Choose Option: ")
    print("Press 1 to manipulate image")
    print("Press 2 to filter image")
    choice = input("Enter your option: ")

    if choice == "1":
        print("Choose one option")
        print("1. Add Borders")
        print("2. Roatate Image")
        print("3. Scale Image")
        print("4. Skew Image")
        print("5. Translate Image")
        img_mani = input("Enter your choice: ")
        
        if img_mani == "1":
            border = int(input("Enter border width in 10 multiples: "))
            imgObj.img_border(border)

        elif img_mani == "2":
            deg = int(input("By how many degree you have to rotate: "))
            imgObj.img_rotate(deg)

        elif img_mani == "3":
            scale = input("Enter Scale example 1x or 2x in X and Y axis: ")
            imgObj.img_scale(scale)

        elif img_mani == "4":
            skew_choice = input("How you want to skew it horizontal or vertical: ")
            imgObj.img_perspective(skew_choice)

        elif img_mani == "5":
            translate = input("Enter how much to want to translate in X and Y Direction: ")
            imgObj.img_translation(translate)

    elif choice == "2":
        print("Select one the options to filter: ")
        print("1. Convert Into Grey Image: ")
        print("2. Convert Into Blue Image: ")
        print("3. Convert Into Green Image: ")
        print("4. Convert Into Red Image: ")
        print("5. Merge two images: ")
        print("6. Blend Two Images: ")
        print("7. Image thresholding: ")
        print("8. Smoothing Image")
        filter_choice = input("Enter Your Choice: ")

        if filter_choice == "1":
            imgObj.grey_img()

        elif filter_choice == "2":
            imgObj.blue_img()

        elif filter_choice == "3":
            imgObj.green_img()

        elif filter_choice == "4":
            imgObj.red_img()

        elif filter_choice == "5":
            second_image = input("Enter Second Path For Second Image: ")
            imgObj.merge_image(second_image)

        elif filter_choice == "6":
            second_image = input("Enter Second Path For Second Image: ")
            imgObj.blend_image(second_image)

        elif filter_choice == "7":
            imgObj.img_threshold()

        elif filter_choice == "8":
            print("Select Smoothing Method: ")
            print("1. Averaging")
            print("2. Gaussian Filtering")
            method_number = input("Enter serial no for selecting method")
            imgObj.img_blur(method_number)
        
            
    
if __name__ == "__main__":
    main()
    
```
